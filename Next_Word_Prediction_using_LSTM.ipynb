```python
tokens = word_tokenize(document.lower())
```

```python
tokens
```

```python
vocab = {"unk" : 0 }
for token in Counter(tokens).keys():
  if token not in vocab:
    vocab[token] = len(vocab)
vocab
```

```python
len(vocab)
```

```python
def text_to_indices(sentence, vocab):
  num_sentence = []
  for word in word_tokenize(sentence):
    word_lower = word.lower()
    if word_lower in vocab:
      num_sentence.append(vocab[word_lower])
    else:
      num_sentence.append(vocab['unk'])
  return num_sentence
```

```python
input_sentence = []
for sentence in document.split("."):
  input_sentence.append(text_to_indices(sentence, vocab))
input_sentence
```

```python
training_sequence = []
for sentence in input_sentence:
  for i in range(1, len(sentence)):
    training_sequence.append(sentence[:i+1])
```

```python
len(training_sequence)
```

```python
len_list = []
for sequence in training_sequence:
  len_list.append(len(sequence))
max_len = max(len_list)
```

```python
padded_sequence = []

for sequence in training_sequence:
  padded_sequence.append([0]* (max_len - len(sequence)) + sequence)
```

```python
len(padded_sequence[6])
```

```python
padded_sequence = torch.tensor(padded_sequence, dtype = torch.long)
```

```python
padded_sequence
```

```python
X = padded_sequence[:,:-1]
y = padded_sequence[:,-1]
```

```python
class CustomDataset(Dataset):

  def __init__(self,X, y):
    self.X = X
    self.y = y

  def __len__(self):
    return self.X.shape[0]

  def __getitem__(self, idx):
    return self.X[idx], self.y[idx]
```

```python
dataset = CustomDataset(X,y)
```

```python
dataloader = DataLoader(dataset, batch_size = 32, shuffle = True)
```

```python
class LSTMNN(nn.Module):

  def __init__(self, vocab_size):
    super().__init__()
    self.embedding = nn.Embedding(vocab_size, 100)
    self.LSTM = nn.LSTM(100, 150, batch_first = True)
    self.fc = nn.Linear(150, vocab_size)

  def forward(self, x):
    embedded = self.embedding(x)
    intermediate_hidden_states, (final_hidden_state, final_cell_state) = self.LSTM(embedded)
    output = self.fc(final_hidden_state.squeeze(0))
    return output
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
model = LSTMNN(len(vocab)).to(device)
```

```python
epochs = 50
learning_rate = 0.001
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr = learning_rate)
```

```python
for epoch in range(epochs):
  total_loss = 0

  for X,y in dataloader:
    X, y = X.to(device), y.to(device)

    optimizer.zero_grad()

    output = model(X)

    loss = criterion(output, y)

    loss.backward()

    optimizer.step()

    total_loss += loss.item()

  print(f"Epoch: {epoch+1}, Loss: {total_loss/len(dataloader)}")
```

```python
# prediction

def prediction(model, vocab, text):

  numerical_text = text_to_indices(text, vocab)

  # padding
  padded_text = torch.tensor([0] * (max_len - len(numerical_text)) + numerical_text, dtype=torch.long).unsqueeze(0)

  # send to model
  output = model(padded_text.to(device))

  # predicted index
  value, index = torch.max(output, dim=1)

  # merge with text
  return text + " " + list(vocab.keys())[index]
```

```python
prediction(model, vocab, "The course follows a monthly")
```

```python
testloader = DataLoader(dataset, batch_size = 32, shuffle = False)
```

```python
def calculate_accuracy(model, dataloader):

  total = 0
  correct = 0
  model.eval()

  with torch.no_grad():
    for x,y in testloader:
      x, y = x.to(device), y.to(device)

      output = model(x)

      _, predicted = torch.max(output, dim = 1)
      correct+= (predicted == y).sum().item()

      total += y.size(0)

  accuracy = (correct/total) *100

  return accuracy
```

```python
accuracy = calculate_accuracy(model, testloader)
print(f"accuracy : {accuracy}")
```
